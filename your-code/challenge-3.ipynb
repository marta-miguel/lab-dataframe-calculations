```python
# your code here

# First, identify two groups of customers from the data set. 
# The first group is VIP Customers whose aggregated expenses at your global chain stores are above the 95th percentile (aka. 0.95 quantile). 
# The second group is Preferred Customers whose aggregated expenses are between the 75th and 95th percentile.

#1.

cust = orders.groupby("CustomerID").agg({"amount_spent":"sum"})

#print(cust)

#2.

# print(cust.max())
# print(cust.min())

amt = np.percentile(cust["amount_spent"], [75,95])
#print(amt)

#3.

labels = ["Regular","Preferred","VIP"]

## Bins:
# VIP = 280206.02 to 5840.182
# Preferred = 5840.182 to 1661.64
# Regular = 1661.64 to 0

bins = pd.cut(cust["amount_spent"], [0, 1661.64, 5840.182, 280206.02], labels = labels)
bins.value_counts()

cust["Cust type"] = bins

display(cust)

bins.value_counts()

```

## Q2: How to identify which country has the most VIP Customers?

```python
# # your code here

cust["Country"] = orders["Country"]

#display(cust)

vip = cust.groupby(["Cust type", "Country"]).agg({"Country":"count"})

display(vip)

# United Kingdom is the country with the most VIP customers
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?
